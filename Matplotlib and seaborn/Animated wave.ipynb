## Sine wave

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating data for the sine wave
x = np.linspace(0, 20, 100)
y = np.sin(x)

# Creating the plot
plt.figure(figsize=(6, 5))
plt.plot(x, y, label="Sine Wave")

# Adding titles and labels
plt.title("Sin Wave")
plt.xlabel("X-axis Data (Radians)")
plt.ylabel("Y-axis Data (Amplitude)")

# Optional but useful for visualization
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generating data for the sine wave
x = np.linspace(0, 20, 100)
y = np.sin(x)  # Define the initial y values

# Setting up the plot
plt.figure(figsize=(6, 5))
line, = plt.plot(x, y, label="Sine Wave")  # Store the line object for animation

# Setting labels and title
plt.title("Sine Wave")
plt.xlabel("X-axis Data (Radians)")
plt.ylabel("Y-axis Data (Amplitude)")
plt.legend()
plt.grid(True)

# Function to update the sine wave for each frame
def update(frame):
    # Shift the sine wave horizontally by a small phase based on the frame
    y = np.sin(x + frame / 10.0)  # Update y with a phase shift
    line.set_ydata(y)  # Update the y-data of the line to create animation
    return line,

# Creating the animation and save
ani = FuncAnimation(plt.gcf(), update, frames=np.arange(0, 100), interval=50, blit=True)
ani.save("sine_wave_animation.gif", writer="pillow")

plt.tight_layout()
plt.show()
```

## **Tan wave**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generating data for the tangent wave
x = np.linspace(0, 20, 100)

# Setting up the plot
plt.figure(figsize=(6, 5))  # Correctly set the figure size
line, = plt.plot(x, np.tan(x), label="Tangent Wave", color="purple")  # Plot initial tangent wave

# Setting labels and title
plt.title("Tangent Wave")
plt.xlabel("X-axis Data (Radians)")
plt.ylabel("Y-axis Data (Amplitude)")
plt.legend()
plt.grid(True)

# Limiting the y-axis to avoid extreme tangent values
plt.ylim(-5, 5)  # Set y-axis limits directly with plt.ylim

# Function to update the tangent wave for each frame
def update(frame):
    # Shift the tangent wave horizontally by a small phase based on the frame
    y = np.tan(x + frame / 10.0)  # Update y with a phase shift
    line.set_ydata(y)  # Update the y-data of the line to create animation
    return line,

# Creating the animation
ani = FuncAnimation(plt.gcf(), update, frames=np.arange(0, 100), interval=50, blit=True)

# Save the animation as a GIF 
ani.save("tangent_wave_animation.gif", writer="pillow")

# Show the animated plot
plt.tight_layout()
plt.show()
```

## **Sine, cos, tan wave**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating x values and computing y values for each function
x = np.linspace(0, 20, 100)
y = np.sin(x)       # Sine function
new = np.cos(x)     # Cosine function
update = np.tan(x)  # Tangent function

# Creating the figure and plotting each function
plt.figure(figsize=(6, 5))
plt.plot(x, y, label='Sin(x)', color='blue')
plt.plot(x, new, label='Cos(x)', color='red')
plt.plot(x, update, label='Tan(x)', color='black')

# Adding title and labels
plt.title("Trigonometric Functions")
plt.xlabel("X-axis Data (Radians)")
plt.ylabel("Y-axis Data (Amplitude)")

# Adding a legend and a grid for better visualization
plt.legend()
plt.grid(True)

# Limiting the y-axis to avoid extreme values from the tangent function
plt.ylim(-5, 5)  # Adjust y-axis limits to keep the tangent function manageable

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generating x values
x = np.linspace(0, 20, 100)

# Calculating y values for each function
y_sin = np.sin(x)       
y_cos = np.cos(x)      
y_tan = np.tan(x)      

# Setting up the plot
plt.figure(figsize=(6, 5))
line_sin, = plt.plot([], [], label='Sin(x)', color='blue')
line_cos, = plt.plot([], [], label='Cos(x)', color='red')
line_tan, = plt.plot([], [], label='Tan(x)', color='black')

# Adding title, labels, legend, and grid
plt.title("Animated Trigonometric Functions")
plt.xlabel("X-axis Data (Radians)")
plt.ylabel("Y-axis Data (Amplitude)")
plt.legend()
plt.grid(True)

# Setting y-axis limits to prevent tangent values from distorting the plot
plt.ylim(-5, 5)
plt.xlim(0, 20)

# Function to update the lines for each frame
def update(frame):
    # Update the x and y data of each line to create the animation effect
    line_sin.set_data(x[:frame], y_sin[:frame])  # Update sine line data
    line_cos.set_data(x[:frame], y_cos[:frame])  # Update cosine line data
    line_tan.set_data(x[:frame], y_tan[:frame])  # Update tangent line data
    return line_sin, line_cos, line_tan

# Creating the animation
ani = FuncAnimation(plt.gcf(), update, frames=len(x), interval=50, blit=True)

# Save the animation as a GIF
ani.save("sine_cosine_tangent_animation.gif", writer="pillow")

# Show the animated plot
plt.tight_layout()
plt.show()
```

## **Bar plot**

```python
import matplotlib.pyplot as plt

# Data for the bar plot
x = ["Cats", "Dogs", "Horse"]  # Categories for x-axis
y = [10, 3, 25]  # Values for y-axis

# Creating the plot
plt.figure(figsize=(6, 5))  # Set the size of the plot
plt.bar(x, y, color='blue')  # Create the bar plot, setting color to blue

# Adding title and labels
plt.title("Bar Plot")  # Title of the plot
plt.xlabel("Animal Types")  # Label for x-axis
plt.ylabel("Count")  # Label for y-axis

# Optional features for better visualization
plt.legend(["Count"])  # Adding a legend
plt.tight_layout()  # Adjust layout to prevent clipping
plt.grid(False)  # Turn off grid if not needed
plt.show()  # Display the plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data for the bar plot
x = ["Cats", "Dogs", "Horse"]
y = [10, 3, 25]

# Creating the figure and axis objects
fig, ax = plt.subplots(figsize=(6, 5))

# Create a bar plot with initial height of 0 (for animation effect)
bars = plt.bar(x, [0, 0, 0], color='blue')

# Adding title and labels
ax.set_title("Bar Plot (Animated)")
ax.set_xlabel("Animal Types")
ax.set_ylabel("Count")
ax.set_ylim(0, max(y) * 1.1)  # Set y-limit to ensure space for full bar height

# Function to update the height of the bars for animation
def update(frame):
    for bar, target_height in zip(bars, y):
        # Update the bar's height linearly until it reaches the target height
        bar.set_height(target_height * frame / 100)  # Gradually reach target height
    return bars

# Creating the animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 101), interval=50, blit=False)

# Save the animation as a GIF using Pillow
ani.save("Bar_plot_animation.gif", writer="pillow")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Creating a dataset with categories and corresponding values
datasets = pd.DataFrame({
    "X": ["Cats", "Dogs", "Horse", "Human", "Car", "Bus", "Biman", "Rocket", "Space", "Cycle"],
    "Y": [1, 3, 2, 1, 3, 2, 1, 3, 2, 4]
})

# Setting up the figure and seaborn plot
fig, ax = plt.subplots(figsize=(8, 5))
sns.set_theme(style="whitegrid")
bar_plot = sns.barplot(x="X", y="Y", data=datasets, palette="viridis", ax=ax)

# Initialize the heights of bars to zero
for bar in bar_plot.patches:
    bar.set_height(0)

# Function to animate the bar plot
def animate(frame):
    # Gradually increase the height of each bar based on the frame count
    for i, bar in enumerate(bar_plot.patches):
        final_height = datasets["Y"][i]
        bar.set_height(min(final_height * frame / 100, final_height))  # Ensure height doesn’t exceed final

# Create the animation
ani = FuncAnimation(fig, animate, frames=100, interval=50, repeat=False)
ani.save("animated_barplot.gif", writer="pillow", fps=20)

# Show the animated plot
plt.title("Animated Bar Plot")
plt.xlabel("Categories (X-axis)")
plt.ylabel("Values (Y-axis)")
plt.tight_layout()
plt.show()
```

## **Pie plot**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data for the pie chart
labels = ["Cats", "Dogs", "Horse"]
sizes = [10, 3, 25]
colors = ['blue', 'green', 'red']

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Initial explode values (all set to zero, so no slice is exploded at the start)
explode = [0, 0, 0]

# Function to update the pie chart for each frame
def update(frame):
    # Gradually increase the explode values for the animation effect
    explode = [0.1 * (frame / 100) for _ in sizes]  # Dynamically increase each slice explosion
    ax.clear()  # Clear the plot to redraw
    ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    ax.set_title("Animated Pie Chart")
    return ax,

# Create the animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 101), interval=50, blit=False)

# Save the animation as a GIF using Pillow as the writer
ani.save("pie_chart_animation.gif", writer="pillow")

# Display the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Data for the pie chart
labels = ["Space", "Rocket", "Biman", "Bus", "Car", "Human", "Horse", "Dogs", "Cats", "Cycle"]
sizes = [15, 10, 5, 5, 10, 5, 10, 10, 5, 25]  # Sample percentages for each category
colors = ['yellowgreen', 'lightcoral', 'orange', 'lightgreen', 'cyan', 'pink', 'gold', 'salmon', 'lime', 'orchid']

# Initialize figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Initial explode values (starting with no explosion)
explode = [0] * len(sizes)

# Function to update the pie chart for each frame
def update(frame):
    # Gradually increase the explode values to animate
    explode = [0.1 * (frame / 100) for _ in sizes]  # Dynamically adjust the explode value
    ax.clear()  # Clear the axis to redraw
    ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    ax.set_title("Animated Pie Chart")
    return ax,

# Create the animation
ani = FuncAnimation(fig, update, frames=np.arange(0, 101), interval=50, blit=False)

# Save the animation as a GIF using Pillow
ani.save("animated_pie_chart.gif", writer="pillow")

# Display the animation
plt.show()
```

## **Hist plot**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import seaborn as sns

# Data for the histogram
data = np.random.normal(0, 1, 1000)  # Generate 1000 random points from a normal distribution
bins = 30  # Number of bins in the histogram

# Setting up the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(-4, 4)
ax.set_ylim(0, 100)
ax.set_title("Animated Histogram")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

# Initialize an empty histogram
n, _, _ = ax.hist([], bins=bins, color="blue", edgecolor="black", alpha=0.7)

# Function to update the histogram for each frame
def update(frame):
    ax.clear()  # Clear the previous histogram
    ax.set_xlim(-4, 4)
    ax.set_ylim(0, 100)
    ax.set_title("Animated Histogram")
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")

    # Use only the first `frame` data points for the current histogram
    data_so_far = data[:frame]
    ax.hist(data_so_far, bins=bins, color="blue", edgecolor="black", alpha=0.7)

# Create the animation and save
ani = FuncAnimation(fig, update, frames=np.arange(10, len(data) + 10, 10), interval=100)
ani.save("animated_histogram.gif", writer="pillow")

plt.show()
```
