```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set up figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 10)  # Set x-axis limits to observe the wave moving
ax.set_ylim(-1.5, 1.5)  # Set y-axis limits for wave amplitude
line, = ax.plot([], [], lw=2)  # Initialize an empty plot for the wave

# Parameters for the wave
frequency = 2  # Wave frequency
wavelength = 2  # Wave wavelength
speed = 1  # Speed of the wave

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Animation function
def animate(t):
    x = np.linspace(0, 10, 1000)
    y = np.sin(2 * np.pi * (frequency * (x / wavelength) - speed * t))
    line.set_data(x, y)
    return line,

# Create the animation
ani = FuncAnimation(fig, animate, frames=200, init_func=init, blit=True, interval=20)

plt.xlabel("Position")
plt.ylabel("Amplitude")
plt.title("Seismic Wave Propagation Animation")
plt.show()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Set up figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)
line, = ax.plot([], [], lw=2)

# Parameters for the wave
frequency = 2
wavelength = 2
speed = 1

def init():
    line.set_data([], [])
    return line,

def animate(t):
    x = np.linspace(0, 10, 1000)
    y = np.sin(2 * np.pi * (frequency * (x / wavelength) - speed * t))
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, animate, frames=200, init_func=init, blit=True, interval=20)

# Save the animation as a GIF
ani.save('wave_animation.gif', writer=PillowWriter(fps=30))

plt.xlabel("Position")
plt.ylabel("Amplitude")
plt.title("Seismic Wave Propagation Animation")
plt.show()
```
