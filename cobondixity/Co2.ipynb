```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Sample Data for Weight vs Fuel Economy
weight = np.array([2.1, 2.3, 2.5, 2.7, 2.9, 3.0, 3.1, 3.3, 3.5, 3.7, 3.9, 4.0])
fuel_economy = np.array([35, 33, 31, 30, 28, 27, 25, 23, 22, 21, 19, 18])

# Setup the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(1.5, 4.5)
ax.set_ylim(10, 40)
ax.set_title("Weight vs Fuel Economy of Cars")
ax.set_xlabel("Weight (in Thousand Pounds)")
ax.set_ylabel("Fuel Economy (in Miles/Gallon)")

# Initialize an empty line plot for animation
line, = ax.plot([], [], 'o', color='b')

# Animation function
def animate(i):
    # Update the data for the line plot by adding points up to index i
    line.set_data(weight[:i], fuel_economy[:i])
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(weight)+1, interval=300, blit=True)

# Display the animation
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Sample data (replace with actual CO2 emission data)
years = np.arange(1850, 2021)
fossil_fuel_emission = np.cumsum(np.random.rand(len(years)) - 0.5)  # Simulating some data
land_use_change_emission = np.cumsum(np.random.rand(len(years)) - 0.5) * 0.1  # Simulating some data

# Setup the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(1850, 2020)
ax.set_ylim(0, 40)
ax.set_title("Global CO₂ Emission")
ax.set_xlabel("Year")
ax.set_ylabel("CO₂ Emission (In Billion Tonnes)")

# Initialize two line objects for Fossil Fuel and Land Use Change emissions
line_fossil, = ax.plot([], [], lw=2, label='Fossil Fuel', color='blue')
line_land_use, = ax.plot([], [], lw=2, label='Land Use Change', color='green')
ax.legend()

# Animation function
def animate(i):
    # Updating fossil fuel emission line with data up to the ith index
    line_fossil.set_data(years[:i], fossil_fuel_emission[:i])
    # Updating land use change emission line with data up to the ith index
    line_land_use.set_data(years[:i], land_use_change_emission[:i])
    return line_fossil, line_land_use

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(years), interval=50, blit=True)

# Display the animation
plt.show()
```
